```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
read = '/content/drive/My Drive/Labwork/as depo.csv'
x = pd.read_csv(read)

gateV = pd.DataFrame(x, columns=['Gate Vg (V)'])
drainI = pd.DataFrame(x, columns=['Drain Id (A)'])
plt.title("as Depo",fontsize=25)
plt.xlabel("Gate V")
plt.ylabel("Drain I")
plt.plot(gateV,np.log(drainI))
plt.show()
```
